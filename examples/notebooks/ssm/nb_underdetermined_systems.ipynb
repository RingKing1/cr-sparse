```python
import jax
import jax.numpy as jnp
import numpy as  np
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig_params = dict(figsize=(8,8), dpi= 100, facecolor='w', edgecolor='k')
```

```python
def draw_axes(ax):
    ax.set_xlim([-6, 6])
    ax.set_ylim([-6, 6])
    ax.set_xticks(np.arange(-6, 7, 1))
    ax.set_yticks(np.arange(-6, 7, 1))

    ax.grid(color='red', linestyle='dotted', linewidth=0.7, alpha=0.1)

    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.axvline(0, linewidth=0.8, alpha=0.3)
    ax.axhline(0, linewidth=0.8, alpha=0.3)
    
    ax.scatter([0],[0], marker='.')

    
```

```python
def draw_system(ax):
    ax.axline([0,3], [4,0]);
    point_style=dict(s=100, c='brown')

    ax.scatter(0, 3, marker='s', **point_style)
    ax.scatter(4, 0, marker='s', **point_style)
```

```python
fig, ax =plt.subplots(**fig_params)

draw_axes(ax)
draw_system(ax)
ax.set_title('Solution of $3x_1 + 4 x_2 = 12$');    
```

```python
def draw_circles(ax):
    radii = [0.2, 0.5, 1, 1.5, 2, 3, 3.5, 4]
    for r in radii:
        circle = plt.Circle((0, 0), r, fill=False)
        ax.add_artist(circle)
    # solution ball
    circle = plt.Circle((0, 0), 2.4, fill=False, color='red')
    ax.add_artist(circle)
    # solution point
    point_style=dict(marker='+', s=100, c='red')
    ax.scatter(1.44, 1.92,  **point_style)
    pass
```

```python
fig, ax =plt.subplots(**fig_params)
draw_axes(ax)
draw_system(ax)
draw_circles(ax)
```
